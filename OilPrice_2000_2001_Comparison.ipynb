```python
# Import dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import json
import time
from pandas.io.json import json_normalize

# Use seaborn style defaults and set the default figure size
import seaborn as sns
sns.set(rc={'figure.figsize':(12, 4)})

# New dependency to format the dates
import matplotlib.dates as mdates
```

```python
# Get data from Alpha Vantage using API request

# Pass the URL and state parameters
url = "https://www.alphavantage.co/query"
params = {
        "function": "TIME_SERIES_DAILY_ADJUSTED",
        "symbol": "CL",
        "outputsize": "full",
        "apikey": "ac9ba10aa04a7d"}


# Get Json response
response = requests.get(url, params=params)
data = response.json()

# Define the data columns we want passed
def df_alpha_response(data,
         columns=['Open', 'High', 'Low', 'Close', 'AdjClose', 'Volume']):

    def alpha_response(d):
        # convert the response into datetimerecords that can be parsed by Pandas
        for dt, prec in d['Time Series (Daily)'].items():
            r = { 'date': dt}
            r.update(prec)
            yield r
    #Create dataframe from response           
    oil_df = pd.DataFrame(alpha_response(data))
    # rename the columns    
    oil_df = oil_df.rename(columns={'1. open': 'Open',
                            '2. high': 'High',
                            '3. low': 'Low',
                            '4. close': 'Close',
                            '5. adjusted close': 'AdjClose',
                            '6. volume': 'Volume'})
    #Crate dataframe of results with new column names
    oil_df['date'] = pd.to_datetime(oil_df['date'])
    # and set date as the index
    oil_df.set_index('date', inplace=True)
    # Sort object by labels (along an axis)
    oil_df.sort_index(inplace=True)
    # extract the default columns
    oil_df = oil_df[columns]
    return oil_df

# Reset index
oil_df = df_alpha_response(data)
oil_df.reset_index(level=0, inplace=True)
oil_df.head(3)
```

```python
# Convert date object to datetime dtype
oil_df['date'] = pd.to_datetime(oil_df['date'])
oil_df.dtypes
```

```python
# Set the date an index
oil_df = oil_df.set_index('date')
```

```python
# Add new columns with year, month, and weekday name to the dataframe
oil_df['Year'] = oil_df.index.year
oil_df['Month'] = oil_df.index.month
oil_df['Weekday Name'] = oil_df.index.weekday_name
oil_df.head(3)
```

```python
#For the column values to be used in the plots, we MUST convert them into numeric values first.
# the date was converted into a datetime object after creating the dataframe
oil_df.Open=pd.to_numeric(oil_df.Open)
oil_df.High=pd.to_numeric(oil_df.High)
oil_df.Low=pd.to_numeric(oil_df.Low)
oil_df.Close=pd.to_numeric(oil_df.Close)
oil_df.AdjClose=pd.to_numeric(oil_df.AdjClose)
oil_df.Volume=pd.to_numeric(oil_df.Volume)
oil_df.Year=pd.to_numeric(oil_df.Year)
oil_df.Month=pd.to_numeric(oil_df.Month)
```

## Weekly Line Plot - Crude Oil Stocks Comparison 

```python
# Visually display side by side the two line plots
# The top plot analyses a short date range, from July 01 through September 10, 2001

#Using the dataframes generated above we'll create a weekly line plot
fig, ax = plt.subplots()
ax.plot(oil_df.loc['2001-07-01':'2001-09-10', 'Close'], 
        marker='o', linestyle='-')
ax.set_ylabel('Crude Oil - Closing Prices ($)')
ax.set_title('Line Plot - Weekly Crude Oil Stocks 07/01 - 09/01')

# Set x-axis major ticks to weekly interval, on Mondays 
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MONDAY))

# Format x-tick labels as 3-letter month name and day number
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'));

# Format y-tick labels with dollar symbol and decimals if applicable
ax.yaxis.set_major_formatter(plt.FuncFormatter('${:.0f}'.format))

plt.savefig("lineWeekly2001.png", dpi=150)
plt.show()

#-----------------------------------------------------------------------------------
# This second plot analyses one+ year, from July 2000 through September 10, 2001

fig, ax = plt.subplots()
ax.plot(oil_df.loc['2000-07-01':'2001-09-10', 'Close'], 
        marker='o', linestyle='-')
ax.set_ylabel('Crude Oil - Closing Prices ($)')
ax.set_title('Line Plot - Weekly Crude Oil Stocks 07/00 - 09/01')

# Format x-tick labels as 3-letter month name and day number
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y %b %d'));

# Format y-tick labels with dollar symbol and decimals if applicable
ax.yaxis.set_major_formatter(plt.FuncFormatter('${:.0f}'.format))

plt.savefig("LineWeekly_2000_2001.png", dpi=150)
plt.show()
```

These two plots shows the stability or lack the oil stocks using 2 different date ranges. Short one, from July 2001 - September 2001. And the larger one from July 2000 to September 10, 2001. The plot method used here was selecting the date range from the dataframe by accessing a group of rows and columns by labels with df.loc

## Plot with Resampling and Rolling Mean

We will resample the data to a weekly mean time series

```python
# Specify the data columns we want to include
data_columns = ['Open', 'Close', 'AdjClose', 'Volume']

# Resample to weekly frequency, aggregating with mean
oil_df_weekly_mean = oil_df[data_columns].resample('W').mean()
oil_df_weekly_mean.head(3)
```

```python
print(oil_df.shape[0])
print(oil_df_weekly_mean.shape[0])
```

#### Line Plot with Resampling and Rolling Mean Functions - Crude Oil 07/2000 - 09/2001

```python
# Create a new dataframe centered on the 7-day rolling mean
oil_df_7day = oil_df[data_columns].rolling(7, center=True).mean()
oil_df_7day.head(6)
```

```python
# Start and end of the date range to extract
start, end = '2001-07-01', '2001-09-10'

# Plot daily resampled to weekly, and 7-day rolling mean (moving average) time series together
fig, ax = plt.subplots()

# Plotting the daily mean
ax.plot(oil_df.loc[start:end, 'Close'],
        marker='.', linestyle='-', linewidth=0.5, label='Daily Mean')

# Using the resampling dataframe to illustrate the weekly mean
ax.plot(oil_df_weekly_mean.loc[start:end, 'Close'],
        marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')

# Using the rolling mean dataframe to illustrate the moving average
ax.plot(oil_df_7day.loc[start:end, 'Close'],
        marker='.', linestyle='-', label='7 Day Rolling Mean')

ax.set_ylabel('2001 Crude Oil - Closing Prices')
ax.set_title('Crude Oil Stock Prices from 07/01 - 09/01')
ax.legend()

# Set x-axis major ticks to weekly interval, on Mondays 
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MONDAY))

# Format x-tick labels as 3-letter month name and day number
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'));

# Format y-tick labels with dollar symbol and decimals if applicable
ax.yaxis.set_major_formatter(plt.FuncFormatter('${:.0f}'.format))

# Rotate the labels at 90 angle when overlapping occur
plt.xticks(rotation=90)

# Save a png image of the plot
plt.savefig("2001rollingmean.png", dpi=150)
plt.show()
#-------------------------------------------------------------------------------

# Start and end of the date range to extract
start, end = '2000-07-01', '2001-09-11'

# Plot daily resampled to weekly, and 7-day rolling mean (moving average) time series together
fig, ax = plt.subplots()

# Plotting the daily mean
ax.plot(oil_df.loc[start:end, 'Close'],
        marker='.', linestyle='-', linewidth=0.5, label='Daily Mean')

# Using the resampling dataframe to illustrate the weekly mean
ax.plot(oil_df_weekly_mean.loc[start:end, 'Close'],
        marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')

# Using the rolling mean dataframe to illustrate the moving average
ax.plot(oil_df_7day.loc[start:end, 'Close'],
        marker='.', linestyle='-', label='7 Day Rolling Mean')

ax.set_ylabel('2000 - 2001 Crude Oil - Closing Prices')
ax.set_title('Crude Oil Stock Prices from 07/00 - 09/01')
ax.legend()

# Format x-tick labels as 3-letter month name and day number
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b, %d'));

# Format y-tick labels with dollar symbol and decimals if applicable
ax.yaxis.set_major_formatter(plt.FuncFormatter('${:.0f}'.format))

# Rotate the labels at 90 angle when overlapping occur
plt.xticks(rotation=90)

# Save a png image of the plot
plt.savefig("00_01rollingmean.png", dpi=150)
plt.show()
```

This plot includes resampling (downsampling) of the daily data to weekly data.
Rolling means (or moving averages) have also been used, which are used to smooth out short-term fluctuations in time series data and highlight long-term trends.

#### 365 Day Moving Average Trends  - Crude Oil Stocks 2000-2019 

```python
# The min_periods=360 argument accounts for a few isolated missing days in the
# Open and AdjClose Prices time series

oil_df_365d = oil_df[data_columns].rolling(window=365, center=True, min_periods=360).mean()
```

```python
# Start and end of the date range to extract
start, end = '2000', '2019'

# Plot daily, 7-day rolling mean, and 365-day rolling mean time series
fig, ax = plt.subplots()

# Plotting the daily mean
ax.plot(oil_df.loc[start:end, 'Close'], marker='.', markersize=2, color='0.6',
        linestyle='None', label='Daily Mean')

# Using the resampling dataframe to illustrate the weekly mean
ax.plot(oil_df_7day.loc[start:end, 'Close'], linewidth=2, label='Weekly Rolling Mean')

# Using the rolling mean dataframe to illustrate the moving average
ax.plot(oil_df_365d.loc[start:end, 'Close'], color='0.2', linewidth=3,
        label='Trend (360 Day Rolling Mean)')

# Set x-ticks to yearly interval and add legend and labels
ax.xaxis.set_major_locator(mdates.YearLocator())

# Format y-tick labels with dollar symbol and decimals if applicable
ax.yaxis.set_major_formatter(plt.FuncFormatter('${:.0f}'.format))

ax.legend()
ax.set_ylabel('Oil Stock Prices ($)')
ax.set_title('Price Trend for Crude Oil from 2000 to 2019')
# Rotate the labels at 45 angle when overlapping occur
plt.xticks(rotation=45)
plt.savefig("365mean2000_2019.png", dpi=150)
plt.show();
```

The trend line reveals that oil prices were on a steady climb from 2000 through 2013, with several ups and downs in between. But the sharp decline of 2013 has remained flat since.

# Create new bar plot using the pivot table function

```python
# Select the date range we want to plot (this one will be for the year 2000 to 2001)
df_2000 = oil_df.loc['2000-01-01':'2001-12-31']
```

```python
# We use the pivot table function to pivot the data and place years as columns
oil_Close_00 = pd.pivot_table(df_2000, index = ['Year','Month'], values= ["Close", "Volume"])
```

```python
#This plot shows the pivot data above, while assigning 3 letter months to the x axis and a stacked bar chart
oil_Close_00.pivot_table(index=['Month'], columns='Year', values='Close', aggfunc='mean').plot.bar(stacked=True)
plt.gca().set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.ylabel("Crude Oil - Closing Prices")
plt.title('Bar Plot - Two Year Comparison 2000 - 2001')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.savefig("Bar2000_2001.png", dpi=150)
plt.show()
```
