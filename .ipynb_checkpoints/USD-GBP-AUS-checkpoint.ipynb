```python
# Import dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import json
import time
from pandas.io.json import json_normalize
pd.plotting.register_matplotlib_converters(explicit=True)

# The plots will be style using the seaborm library.
import seaborn as sns

# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(12, 8)})

# Import a new dependency to format the dates
import matplotlib.dates as mdates
```

```python

# Get data from Alpha Vantage using API request

# Pass the URL and state parameters
url = "https://www.alphavantage.co/query"
params = {
        "function": "FX_DAILY",
        "from_symbol": "EUR",
        "to_symbol" : "USD",
        "outputsize": "full",
        "apikey": "GN0RJ89DOWPDQYJR"}


# Get Json response
response = requests.get(url, params=params)
data = response.json()

# Define the data columns we want passed
def df_alpha_response(data,
         columns=['Open', 'High', 'Low', 'Close']):

    def results_alpha_response(d):
        # convert the response into datetimerecords that can be parsed by Pandas
        for dt, prec in d['Time Series FX (Daily)'].items():
            r = { 'date': dt}
            r.update(prec)
            yield r
               
    usd2eur_df = pd.DataFrame(results_alpha_response(data))
    # rename the columns    
    usd2eur_df = usd2eur_df.rename(columns={'1. open': 'Open',
                            '2. high': 'High',
                            '3. low': 'Low',
                            '4. close': 'Close',
                            '5. adjusted close': 'AdjClose',
                            '6. volume': 'Volume'})
    usd2eur_df['date'] = pd.to_datetime(usd2eur_df['date'])
    usd2eur_df.set_index('date', inplace=True)
    usd2eur_df.sort_index(inplace=True)
    # extract the default columns
    usd2eur_df = usd2eur_df[columns]
    return usd2eur_df


# Reset index
usd2eur_df = df_alpha_response(data)
usd2eur_df.reset_index(level=0, inplace=True)
usd2eur_df.head()
```

```python

# Add columns with year, month, and weekday name to the dataframe
usd2eur_df['Year'] = usd2eur_df.index.year
usd2eur_df['Month'] = usd2eur_df.index.month
usd2eur_df['Weekday Name'] = usd2eur_df.index.weekday_name

# Display a random sampling of 5 rows
usd2eur_df.sample(5, random_state=0)
```
